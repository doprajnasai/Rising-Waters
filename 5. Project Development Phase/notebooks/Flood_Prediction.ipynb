```python
# ==========================================
# Handling Outliers using IQR Method
# ==========================================

numerical_columns = [
    "Temp",
    "Humidity",
    "Cloud Cover",
    "ANNUAL",
    "Jan-Feb",
    "Mar-May",
    "Jun-Sep",
    "Oct-Dec",
    "avgjune"
]
for column in numerical_columns:

    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    print(f"\n{column}")
    print(f"Q1 : {Q1}")
    print(f"Q3 : {Q3}")
    print(f"IQR : {IQR}")
    print(f"Lower Limit : {lower_limit}")
    print(f"Upper Limit : {upper_limit}")
    print("-" * 40)
```

```python
# ==========================================
# Applying IQR Capping
# ==========================================

for column in numerical_columns:

    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    dataset[column] = dataset[column].clip(lower=lower_limit,
                                           upper=upper_limit)

print("Outlier capping completed successfully.")
```

```python
# ==========================================
# Box Plots After Outlier Capping
# ==========================================

for column in numerical_columns:

    plt.figure(figsize=(7,2))

    sns.boxplot(x=dataset[column])

    plt.title(f"Box Plot After Capping - {column}")

    plt.xlabel(column)

    plt.show()
```

### Observation

- Outliers were detected using the Interquartile Range (IQR) method.
- Instead of removing rows, the capping technique was applied.
- Values below the lower limit were replaced with the lower limit.
- Values above the upper limit were replaced with the upper limit.
- This preserves the dataset size while reducing the influence of extreme values.

```python
# ==========================================
# Creating Input Features (X) and Target (y)
# ==========================================

X = dataset.iloc[:, :-1].values

y = dataset.iloc[:, -1].values
```

```python
# ==========================================
# Splitting Dataset into Training and Testing
# ==========================================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=10
)
```

```python
# ==========================================
# Creating StandardScaler Object
# ==========================================

sc = StandardScaler()
```

```python
# ==========================================
# Scaling Training Data
# ==========================================

X_train = sc.fit_transform(X_train)
```
